### $\alpha_b$ == $\alpha_p$

```python
world = World(state_size =50, time =4000, agent=["base", "base"], seed=7, pred_a=[.1,.1], behav_a=[.1,.1],
prediction=["chaos", "chaos"], behavior=["chaos", "chaos"], attention=["static", "static"])

world.create_agents()
world.run()
agents = world.get_agents()
error = world.get_avg_abs_error()
preds = world.get_pred()
priors = world.get_behav_priors()
costs = world.get_costs()
predictability = world.get_predictability()
concurrence = world.get_concurrence()
tru_dif = world.get_tru_dif()

#world.print_results()
time = [t for t in range(len(costs[0]))]


plt.figure(facecolor='white', figsize=(15, 5)) 
plt.plot(time, error[0], 'g', label=agents[0])
plt.plot(time, error[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.ylabel("Error")
plt.xlabel("Time")
plt.legend()
plt.title("Raw Error Signals")
plt.show()

plt.figure(facecolor='white', figsize=(15, 5)) 
plt.plot(time, costs[0], 'g', label=agents[0])
plt.plot(time, costs[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.ylabel("Cost")
plt.xlabel("Time")
plt.legend()
plt.title("Costs - Attention-Filtered Error")
plt.show()

plt.figure(facecolor='white', figsize=(15, 5)) 
plt.plot(time, predictability[0], 'g', label=agents[0])
plt.plot(time, predictability[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.ylabel("Predictability")
plt.xlabel("Time")
plt.legend()
plt.title("Predictability of Behavioral Priors")
plt.show()

plt.figure(facecolor='white', figsize=(15, 5)) 
plt.plot(time, tru_dif[0], 'g', label=agents[0])
plt.plot(time, tru_dif[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.ylabel("True Difference")
plt.xlabel("Time")
plt.legend()
plt.title("True Difference in Predictions and Priors")
plt.show()

plt.figure(facecolor='white', figsize=(15, 5)) 
plt.plot(time, concurrence[0], 'b', label="Synchrony")
plt.ylim([0, 1])
plt.ylabel("Synchrony")
plt.xlabel("Time")
plt.legend()
plt.title("Synchrony of Error Between Agents")
plt.show()
```
