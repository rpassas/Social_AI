```python
from world import World 
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sp
import itertools as it
import pandas as pd
```

```python
for state_n in [5, 10, 25, 50, 100, 250, 500]:
    print('state_N:',state_n)
    for mdl_var in [1, 0.5, 0.25, 0.1, 0.05, 0]:
        print('mdl_var:',mdl_var)
        ## SETUP SHEET 1
        agent1 = []
        iteration1 = []
        state_size1 = []
        timesteps = []
        mean_obs_pe = []
        mean_lat_pe = []
        model = []
        memory = []
        control = []
        model_var = []
        learnable = []

        df = pd.DataFrame(columns = ['agent', 'iteration', 'state_size', 'timesteps', 'mean_observed_pe', 'mean_latent_pe', 'model', 'memory', 'control', 'model_var', 'learnable'])
        ## MODEL
        for rep in range(0,1000):
            print('rep:', rep)
            world = World(state_size = state_n, time = 1000, agent=["model_alt", "model_alt"], 
                          memory=[0,0], behav_control=[0,0], model_var=[mdl_var,mdl_var], learnable = [1,1])
            world.create_agents()
            world.run()
            costs = world.get_costs()
            preds = world.get_pred()
            priors = world.get_behav_priors()
            for a in range(0,len(world.agents)): 
                if rep == 0 and a == 0:
                    df = pd.DataFrame({'agent': a,
                                       'iteration': rep,
                                       'state_size': world.state_size,
                                       'timesteps': world.time,
                                       'model': world.type[a],
                                       'memory': world.memory[a],
                                       'control': world.behav_control[a],
                                       'model_var': world.model_var[a],
                                       'learnable':world.learnable[a],
                                       'mean_obs_pe': np.mean(costs[a]),
                                       'mean_lat_pe': np.mean(np.abs(priors[:,1,:] - preds[:,0,:]))}, index = [0])
                    df_time = pd.DataFrame({'agent': a,
                                       'iteration': rep,
                                       'state_size': world.state_size,
                                       'time': [t for t in range(len(costs[0]))],
                                       'model': world.type[a],
                                       'memory': world.memory[a],
                                       'control': world.behav_control[a],
                                       'model_var': world.model_var[a],
                                       'learnable': world.learnable[a],
                                       'obs_pe': costs[a],
                                       'lat_pe': np.mean(np.abs(priors[:,1,:] - preds[:,0,:]), axis=0)})
                    df_priors = pd.DataFrame({'agent': a,
                                       'iteration': rep,
                                       'state_size': world.state_size,
                                       'model': world.type[a],
                                       'memory': world.memory[a],
                                       'control': world.behav_control[a],
                                       'model_var': world.model_var[a],
                                       'learnable': world.learnable[a],
                                       'initial_behav': priors[:,0,0],
                                       'final_behav': priors[:,0,-1],
                                       'initial_pred': preds[:,0,0],
                                       'final_pred': preds[:,0,-1],
                                       'initial_pred_behav_diff': priors[:,1,0] - preds[:,0,0],
                                       'final_pred_behav_diff': priors[:,1,-1] - preds[:,0,-1]})
                elif a == 0:
                    df = pd.concat([df, 
                                  pd.DataFrame({'agent': a,
                                               'iteration': rep,
                                               'state_size': world.state_size,
                                               'timesteps': world.time,
                                               'model': world.type[a],
                                               'memory': world.memory[a],
                                               'control': world.behav_control[a],
                                               'model_var': world.model_var[a],
                                               'learnable':world.learnable[a],
                                               'mean_obs_pe': np.mean(costs[a]),
                                               'mean_lat_pe': np.mean(np.abs(priors[:,1,:] - preds[:,0,:]))}, 
                                               index = [0])])
                    df_time = pd.concat([df_time, 
                                  pd.DataFrame({'agent': a,
                                       'iteration': rep,
                                       'state_size': world.state_size,
                                       'time': [t for t in range(len(costs[0]))],
                                       'model': world.type[a],
                                       'memory': world.memory[a],
                                       'control': world.behav_control[a],
                                       'model_var': world.model_var[a],
                                       'learnable':world.learnable[a],
                                       'obs_pe': costs[a],
                                       'lat_pe': np.mean(np.abs(priors[:,1,:] - preds[:,0,:]), axis=0)})])
                    df_priors = pd.concat([df_priors, 
                                  pd.DataFrame({'agent': a,
                                       'iteration': rep,
                                       'state_size': world.state_size,
                                       'model': world.type[a],
                                       'memory': world.memory[a],
                                       'control': world.behav_control[a],
                                       'model_var': world.model_var[a],
                                       'learnable': world.learnable[a],
                                       'initial_behav': priors[:,0,0],
                                       'final_behav': priors[:,0,-1],
                                       'initial_pred': preds[:,0,0],
                                       'final_pred': preds[:,0,-1],
                                       'initial_pred_behav_diff': priors[:,1,0] - preds[:,0,0],
                                       'final_pred_behav_diff': priors[:,1,-1] - preds[:,0,-1]})])
                elif a == 1:
                    df = pd.concat([df,
                                   pd.DataFrame({'agent': a,
                                               'iteration': rep,
                                               'state_size': world.state_size,
                                               'timesteps': world.time,
                                               'model': world.type[a],
                                               'memory': world.memory[a],
                                               'control': world.behav_control[a],
                                               'model_var': world.model_var[a],
                                               'learnable':world.learnable[a],
                                               'mean_obs_pe': np.mean(costs[a]),
                                               'mean_lat_pe': np.mean(np.abs(priors[:,0,:] - preds[:,1,:]))},
                                               index = [0])])
                    df_time = pd.concat([df_time, 
                                  pd.DataFrame({'agent': a,
                                       'iteration': rep,
                                       'state_size': world.state_size,
                                       'time': [t for t in range(len(costs[0]))],
                                       'model': world.type[a],
                                       'memory': world.memory[a],
                                       'control': world.behav_control[a],
                                       'model_var': world.model_var[a],
                                       'learnable':world.learnable[a],
                                       'obs_pe': costs[a],
                                       'lat_pe': np.mean(np.abs(priors[:,0,:] - preds[:,1,:]), axis=0)})])
                    df_priors = pd.concat([df_priors, 
                                  pd.DataFrame({'agent': a,
                                       'iteration': rep,
                                       'state_size': world.state_size,
                                       'model': world.type[a],
                                       'memory': world.memory[a],
                                       'control': world.behav_control[a],
                                       'model_var': world.model_var[a],
                                       'learnable': world.learnable[a],
                                       'initial_behav': priors[:,0,0],
                                       'final_behav': priors[:,0,-1],
                                       'initial_pred': preds[:,0,0],
                                       'final_pred': preds[:,0,-1],
                                       'initial_pred_behav_diff': priors[:,0,0] - preds[:,1,0],
                                       'final_pred_behav_diff': priors[:,0,-1] - preds[:,1,-1]})])
        # SAVE
        df.to_csv('~/projects/2022/sense_of_should_modeling/'+
                  'MEAN-simulation_statesize-'+str(state_n)+'_var-'+str(mdl_var)+'.csv', index = False)
        df_time.to_csv('~/projects/2022/sense_of_should_modeling/'+
                  'TIMECOURSE-simulation_statesize_'+str(state_n)+'_var-'+str(mdl_var)+'.csv', index = False)
        df_priors.to_csv('~/projects/2022/sense_of_should_modeling/'+
                  'INITIAL-FINAL-PREDS-PRIORS-simulation_statesize_'+str(state_n)+'_var-'+str(mdl_var)+'.csv', index = False)
```
