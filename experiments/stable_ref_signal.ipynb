```python
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
from sandbox.world import World 
```

### One agent has static predictions

```python
world = World(state_size =50, time =500, agent=["base", "base"], seed=7, pred_a= [0,1], prediction=["sigmoid", "sigmoid"], behavior=["sigmoid", "sigmoid"], attention=["static", "static"])

world.create_agents()
world.run()
agents = world.get_agents()
error = world.get_avg_abs_error()
preds = world.get_pred()
priors = world.get_behav_priors()
costs = world.get_costs()
predictability = world.get_predictability()
tru_dif = world.get_tru_dif()

#world.print_results()
time = [t for t in range(len(costs[0]))]
```

```python
plt.figure(facecolor='white') 
plt.plot(time, error[0], 'g', label=agents[0])
plt.plot(time, error[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.legend()
plt.title("Error")
plt.show()
```

```python
plt.figure(facecolor='white') 
plt.plot(time, costs[0], 'g', label=agents[0])
plt.plot(time, costs[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.legend()
plt.title("Costs")
plt.show()
```

```python
plt.figure(facecolor='white') 
plt.plot(time, predictability[0], 'g', label=agents[0])
plt.plot(time, predictability[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.legend()
plt.title("Predictability")
plt.show()
```

```python
plt.figure(facecolor='white') 
plt.plot(time, tru_dif[0], 'g', label=agents[0])
plt.plot(time, tru_dif[1], 'r', label=agents[1])
plt.ylim([0, 1])
plt.legend()
plt.title("True Difference between Priors & Predictions")
plt.show()
```
